```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from math import *
from scipy.signal import *
```

```python
alpha = 1.274
Q = 0.577
f0 = 100 # Hertz
w0 = 2*pi*f0
num = [1]
den = [1/(alpha*w0)**2, 1/(alpha*Q*w0), 1]
filter = lti(num, den)
filter
```

```python
fig, ax = plt.subplots(1,1)
Time, Stepresp = filter.step()
ax.plot(Time, Stepresp)
ax.set_xlabel('time [s]')
ax.set_ylabel('amplitude')
ax.set_title('step response')
ax.grid()
```
